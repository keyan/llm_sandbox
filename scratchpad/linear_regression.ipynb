```python
def fit(X, Y, epochs):
    m = np.random.random()
    b = np.random.random()
    n = len(X)
    lr = 1e-3
    losses = []
    
    for _ in range(epochs):
        y_hat = (m * X) + b
        loss = (1/n) * ((Y - y_hat)**2).sum()
        losses.append(loss)

        # Gradients derived by differentiating: (1/n) * (y - y_hat)**2
        m_grad = (1/n) * (2 * -X * (Y - y_hat)).sum()
        b_grad = (1/n) * (2 * -(Y - y_hat)).sum()
        m -= m_grad * lr
        b -= b_grad * lr

    x = np.linspace(0, 2, 400)  # generate x values from -10 to 10
    y = m * x + b  # calculate corresponding y values

    plt.plot(x, y) # plot fit line
    plt.xlabel('x')
    plt.ylabel('y')
    plt.scatter(X, Y) # plot raw data

    return losses


losses = fit(X, Y, 10000)
```
